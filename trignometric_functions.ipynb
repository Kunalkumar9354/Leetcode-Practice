```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("dark_background") #background colour of graph
```

```python
np.pi
```

```python
np.sin(np.pi/2)
np.sin(np.pi/3)
np.cos(np.pi/2)
np.sin(np.pi/6)
np.tan(0)
```

use matplotlib with numpy

```python
x=np.arange(1,11)
y=np.arange(10,110,10)
plt.figure(figsize=(3,3))
plt.plot(x,y)
plt.show()
```

```python
x_sin=np.arange(0,2*np.pi,0.1)
y_sin=np.sin(x_sin)
print(y_sin)
#plot
plt.figure(figsize=(3,3))
plt.plot(x_sin,y_sin)
plt.title("sin_curve")
plt.show()
```

```python
x_cos=np.arange(0,2*np.pi,0.1)
y_cos=np.cos(x_cos)
print(y_cos)
plt.figure(figsize=(3,3))
plt.plot(x_cos,y_cos)
plt.title("cos_curve")
plt.show()
```

```python
x_tan=np.arange(0,2*np.pi,0.1)
y_tan=np.tan(x_tan)
print(y_tan)
plt.figure(figsize=(3,3))
plt.plot(x_tan,y_tan)
plt.title("tan_curve")
plt.show()
```

```python
x_cot=np.arange(0,2*np.pi,0.1)
y_cot=1/np.tan(x_cot)
plt.figure(figsize=(3,3))
plt.title("cot_curve")
plt.plot(x_cot,y_cot)
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.subplot(2,2,1)
plt.plot(x_sin,y_sin,"r-")
plt.title("sin_curve")

plt.subplot(2,2,2)
plt.plot(x_cos,y_cos,"b-")
plt.title("cos_curve")


plt.subplot(2,2,3)
plt.plot(x_tan,y_tan,"g-")
plt.title("tan_curve")

plt.subplot(2,2,4)
plt.plot(x_cot,y_cot,"w-")
plt.title("cot_curve")

plt.show()
```
