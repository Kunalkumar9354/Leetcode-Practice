```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
  plt.figure(figsize=(8,4))
  plt.subplot(1,2,1)
  sns.histplot(sample)
  plt.subplot(1,2,2)
  stat.probplot(sample,dist="norm",plot=pylab)
  plt.show()
```

```python
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
plot_data(df['sepal_width'])
```
